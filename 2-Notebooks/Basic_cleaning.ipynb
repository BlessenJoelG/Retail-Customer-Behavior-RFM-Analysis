```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
#RAW DATA EXTRACTION 
```

```python
df = pd.read_csv('online_retail_II_reduced.csv')
```

```python
df
```

```python
#RENAMING COLUMNS TO READABLE FORM
```

```python
df.rename(columns={'Customer ID':'Customer_ID'})
```

```python
#CHECKING NULL VALUES 
```

```python
df.isnull().any()
```

```python
#Removed Customer ID with null Values
```

```python
df = df.dropna(subset=['Customer ID']) 
df
```

```python
df.isnull().any()
```

```python
#REMOVING NEGATIVE VALUES
```

```python
df = df[df['Quantity']>0]	#Negative Quantity Removed 
df
```

```python
df = df[df['Price']>0] #Negative Price Removed
df
```

```python
#CHANGING FORMATS
```

```python
df.loc[:, 'InvoiceDate']=pd.to_datetime(df['InvoiceDate']) #Converted InvoiceDate to date fromat.
df
```

```python
#AGAIN CHANGING TO ORIGINAL DATAFRAME 
```

```python
df=df.copy()
df
```

```python
#CREATING A MONETARY VALUE FOR RFM ANALYSIS (feature Engineering)
```

```python
df['TotalAmount']=df['Quantity'] * df['Price'] #Created monetary value of the product for RFM analysis
df
```
