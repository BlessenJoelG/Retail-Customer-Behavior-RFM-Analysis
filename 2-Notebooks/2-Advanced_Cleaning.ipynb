```python
#Checking Outliers
```

```python
#Verification of outliers in Quantity by Inter Quartile Range method (IQR)
```

```python
Q1=df['Quantity'].quantile(0.25)
Q3=df['Quantity'].quantile(0.75)
IQR=Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Lower Bound:", lower_bound)
print("Upper Bound;", upper_bound)
```

```python
df[(df['Quantity'] < lower_bound) |
    (df['Quantity']>upper_bound)].shape
```

```python
#Verification of outliers in Total Amount by Inter Quartile Range method (IQR)
```

```python
Q1=df['TotalAmount (M)'].quantile(0.25)
Q3=df['TotalAmount (M)'].quantile(0.75)
IQR=Q3-Q1
lower_bound_amt = Q1 - 1.5*IQR
upper_bound_amt = Q3 + 1.5*IQR
print("Lower Bound:", lower_bound_amt)
print("Upper Bound;", upper_bound_amt)
```

```python
df[(df['TotalAmount (M)'] < lower_bound) |
    (df['TotalAmount (M)']>upper_bound)].shape
```
